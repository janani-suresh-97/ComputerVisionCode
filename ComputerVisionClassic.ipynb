Code for face detection using a haar cascade classifier

```python
#Import libraries
import cv2
from matplotlib import pyplot as plt
```

```python


# Opening image
img = cv2.imread("/content/person.jpg")

#Conver t images to gray and rgb
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)



liscence_plate_data = cv2.CascadeClassifier('/content/frontal_face.xml')

found = liscence_plate_data.detectMultiScale(img_gray,
                                   minSize =(20, 20))

```

```python
# Don't do anything if there's
# no sign
amount_found = len(found)

if amount_found != 0:
    # There may be more than one
    # object in the image
    for (x, y, width, height) in found:

        # We draw a green rectangle around
        # every recognized sign
        cv2.rectangle(img_rgb, (x, y),
                      (x + height, y + width),
                      (0, 255, 0), 5)

# Creates the environment of
# the picture and shows it
plt.subplot(1, 1, 1)
plt.imshow(img_rgb)
plt.show()
```
